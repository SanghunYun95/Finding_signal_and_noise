```python
import FinanceDataReader as fdr
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
fb = fdr.DataReader('FB', '2015', '2021')
fb.tail()
```

```python
px.line(fb, y='Close', title='FaceBook 일별 종가')
```

```python
px.line(fb, y='Change', title='FaceBook 전일비')
```

```python
px.histogram(fb, x='Change', marginal='violin', title='FaceBook 전일비')
```

```python
fb.head()
```

```python
fig = go.Figure(
    data=[go.Ohlc(
        x=fb.index,
        open=fb['Open'],
        high=fb['High'],
        low=fb['Low'],
        close=fb['Close']
    )]
)
fig.show()
```
