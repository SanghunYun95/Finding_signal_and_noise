```python
import pandas as pd
import cufflinks as cf

df = cf.datagen.ohlcv()
df.head()
```

```python
df[['open', 'high', 'low', 'close']].plot(figsize=(10, 3))
```

```python
# Cufflinks
cf.go_offline(connected=True)
df[['open', 'high', 'low', 'close']].iplot()
```

```python
df[['open', 'high', 'low', 'close']].plot(figsize=(10,5), subplots=True)
```

```python
df[['open', 'high', 'low', 'close']].iplot(subplots=True)
```

```python
qf = cf.QuantFig(df, name='')
qf.iplot()
```

```python
qf.add_bollinger_bands()
qf.iplot()
```

```python
qf.add_sma([30,60], color='red')
qf.iplot()
```

```python
qf = cf.QuantFig(df, name='')
qf.add_rsi(periods=20, color='java')
qf.add_bollinger_bands(periods=20, boll_std=2,
                       colors=['magenta', 'grey'],fill=True)
qf.add_volume()
qf.add_macd()
qf.iplot()
```
